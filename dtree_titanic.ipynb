```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#url = "http://facweb.cs.depaul.edu/mobasher/classes/csc478/Data/titanic-trimmed.csv"
titanic = pd.read_csv("titanic-trimmed.csv")
#titanic = pd.read_csv(url)
titanic.head(10)
```

```python
#getting the count of values which are null in the data set. 
titanic[titanic.age.isnull()].shape
```

```python
#getting the mean of the age - feature.
age_mean = titanic.age.mean()
print age_mean

#fillna - Fill NA/NaN values using the specified method
# http://pandas.pydata.org/pandas-docs/version/0.17.1/generated/pandas.DataFrame.fillna.html
titanic.age.fillna(age_mean, axis=0, inplace=True)

# Return object with labels on given axis omitted where alternately 
#any or all of the data are missing
# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.dropna.html
titanic.dropna(axis=0, inplace=True)
```

```python
titanic.shape
```

```python
titanic.set_index('pid', drop=True, inplace=True)
titanic.head()
```

```python
#Creating dummy variables for categorical featur - all catogorical variable will be converted to dummy variables. 
#titanic_ssf = pd.get_dummies(df['sex']) - if you want to convert only sex to the dummy variable: 
titanic_ssf = pd.get_dummies(titanic) # converting all categorical values into dummy variable for further analysis. 
titanic_ssf.head(10)
```

```python
titanic_names = titanic_ssf.columns.values
titanic_names
```

```python
# assigning survived feature to y
y = titanic_ssf['survived']  
X = titanic_ssf[titanic_names[1:]] 
# assigned rest features to x except "survived". starting from 1 to all. 
X.head()
```

```python
# .T - Transpose index and columns
titanic_ssf.describe().T
```

# Build the training and testing dataset¶


```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=33) 
#random_state - is similar to seed in R
```

## A versatile function to measure performance of a model¶


```python
from sklearn import tree
dt = tree.DecisionTreeClassifier(criterion='entropy')
dt = dt.fit(X_train, y_train)
```

```python
from sklearn import metrics

def measure_performance(X, y, clf, show_accuracy=True, show_classification_report=True, show_confussion_matrix=True):
    y_pred = clf.predict(X)   
    if show_accuracy:
         print "Accuracy:{0:.2f}".format(metrics.accuracy_score(y, y_pred)),"\n" 
            #{} FOR OLD VERSION OF PYTHON WE NEED TO SPECIFY THE POSITION WHERE WE NEED TO PRINT. WE CAN REMOVE IT HERE. 
    if show_classification_report:
        print "Classification report"
        print metrics.classification_report(y, y_pred),"\n"
      
    if show_confussion_matrix:
        print "Confussion matrix"
        print metrics.confusion_matrix(y, y_pred),"\n"
```

```python
from sklearn import metrics
measure_performance(X_test, y_test, dt, show_confussion_matrix=True, 
                    show_classification_report=True)
```
